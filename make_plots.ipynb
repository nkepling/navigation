```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('test.csv')
data.head()
data.columns = ["function","live_lock","seed","inference_time","found_goal","steps"]
data.head()
```

```python

data.groupby("function")["found_goal"].mean().plot(kind='bar')
plt.title("The average number of times the goal was found")
plt.ylabel("Success rate")
plt.xlabel("Function")
```

```python

import seaborn as sns

plt.figure(figsize=(15, 6))
sns.barplot(x='seed', y='steps', hue='function', data=data)

# Add labels and title
plt.xlabel('Seed')
plt.ylabel('Number of Steps')
plt.title('Number of Steps to Goal by Seed and Function, 10x10 grid,(including live lock/failures)')

plt.show()
```

```python

plt.figure(figsize=(15, 6))

found_goal = data[data['found_goal'] == True]
sns.barplot(x='seed', y='steps', hue='function', data=found_goal)

# Add labels and title
plt.xlabel('Seed')
plt.ylabel('Number of Steps')
plt.title('Number of Steps to Goal by Seed and Function, 10x10 grid (only successful runs)')

plt.show()
```

```python
plt.figure(figsize=(15, 6))
sns.barplot(x='seed', y='inference_time', hue='function', data=data)

# Add labels and title
plt.xlabel('Seed')
plt.ylabel('Mean Inference Time per Decision Epoch [s]')
plt.title('Inference Time by Seed and Function, 10x10 Grid,(including failed runs)')

plt.xticks(rotation=90)
# Show the plot
plt.show()
```
